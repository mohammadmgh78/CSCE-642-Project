```python
import numpy as np
import pandas as pd
```

```python
from scipy.stats import norm, expon
num_contact_groups = 2
params = {1:{'mu_p':1, 'scale_p':2, 'mu_s':2, 'scale_s':1, 'mu_a':4},
            2:{'mu_p':1, 'scale_p':2, 'mu_s':2, 'scale_s':1, 'mu_a':3}}


end_time = 540
num_reps = 4
for i in range(1, num_reps+1):
    df = pd.DataFrame()
    for contact_group in range(1, num_contact_groups+1):
        A = expon.rvs(size=5000, loc=0, scale=params[contact_group]['mu_a'])
        A = np.cumsum(A)
        A = A[A<540]
        P = np.exp(norm.rvs(size=len(A), loc=params[contact_group]['mu_p'], scale=params[contact_group]['scale_p']))
        S = np.exp(norm.rvs(size=len(A), loc=params[contact_group]['mu_s'], scale=params[contact_group]['scale_s']))
        df_tmp = pd.DataFrame()
        df_tmp['Arrival Time'] = A
        df_tmp['Handle Time'] = S 
        df_tmp['Patience Time'] = P
        df_tmp['Contact Group'] = contact_group
        df = pd.concat([df, df_tmp])
    
        df.to_csv(f'Simulated_Contact_{i}.csv')   
```
