```python
# for each contact group there should be an object that fits to the historical data and then fits the 
macro_rep = 1
num_reps_simulate = 200
num_contact_groups = 2
macro_rep = macro_rep # so that you knwo they should be the same.
for i in range(1, macro_rep+1):
    for j in range(1, num_reps_simulate+1):
        df_tmp = pd.read_csv(f'data_bo/Back_Office_Dataset.csv')
        df_final_tmp = pd.DataFrame()
        for contact_group in range(1, num_contact_groups+1):
            df_tmp_contact = df_tmp[df_tmp['Contact Group']==contact_group]
            
            df_tmp_contact['BO Handle Time'] = np.random.exponential(scale=2, size=len(df_tmp_contact))*5
            
            df_final_tmp = pd.concat([df_final_tmp, df_tmp_contact])
        df_final_tmp.to_csv(f'data_bo/Back_Office_Sim_True_Macro_{i}_Rep_{j}.csv')  
```
